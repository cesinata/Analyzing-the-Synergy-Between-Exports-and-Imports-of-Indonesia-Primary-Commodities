```python
df_agg = df_1.groupby(['Partner', 'Year']).agg({'Trade Value (US$)': 'sum'})
```

```python
pas = sns.lineplot(y='Trade Value (US$)', x='Year', hue='Partner', data=df_agg, palette='bright')
#ax.grid(True)
plt.xlabel('Tahun')
plt.ylabel('Nilai Ekspor')
plt.title('Ekspor Mobil Indonesia 2016-2020')
plt.show()
```
