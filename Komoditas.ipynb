##KELAPA SAWIT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive

drive.mount('/content/drive')

df_1 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Kelapa Sawit/2016-2020.csv')
df_2 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Kelapa Sawit/2011-2015.csv')
df_3 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Kelapa Sawit/2006-2010.csv')
df_4 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Kelapa Sawit/2001-2005.csv')

```

```python
li = []
li.append(df_1)
li.append(df_2)
li.append(df_3)
li.append(df_4)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
df_agg = frame.groupby(['Partner', 'Year']).agg({'Trade Value (US$)': 'sum'})

df_agg
```

```python
ax = sns.lineplot(y='Trade Value (US$)', x='Year', hue='Partner', data=df_agg, palette='bright')
#ax.grid(True)
plt.xlabel('Tahun')
plt.ylabel('Nilai Ekspor')
plt.title('Ekspor Kelapa Sawit Indonesia 2001-2020')
plt.show()
```
