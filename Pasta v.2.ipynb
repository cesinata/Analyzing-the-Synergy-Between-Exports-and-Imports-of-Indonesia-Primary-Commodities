```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Pasta/2001-2005.csv')
df2 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Pasta/2006-2010.csv')
df3 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Pasta/2011-2015.csv')
df4 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Pasta/2016-2020.csv')
path = '/content/drive/MyDrive/CesAlFem SIC'
```

```python
df5 = df1.append([df2, df3, df4], ignore_index =True)

df5
```

```python
df5['Year'] = df5['Year'].astype('string')
```

```python
df_agg = df5.groupby(['Partner', 'Year']).agg({'Trade Value (US$)': 'sum'})

print(df_agg.to_string())
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

sns.lineplot(data=df_agg, x='Year', y='Trade Value (US$)', hue='Partner')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),fancybox=False, shadow=True, ncol=5,fontsize=12)
plt.grid(False)
plt.yscale('linear')

ax = plt.gca()
#ax.axes.xaxis.set_visible(False)
#ax.axes.yaxis.set_visible(False)

plt.savefig('ekspor pasta.svg', transparent=True)

plt.show()
```
