```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
ps_1 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/tepung2/tepung3.csv')
```

```python
ps_2 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/tepung2/tepung4.csv')
```

```python
pas = ps_1.append([ps_2], ignore_index =True)

pas
```

```python
pas_agg = pas.groupby(['Partner', 'Year']).agg({'Trade Value (US$)': 'sum'})

print(pas_agg.to_string())
```

```python
sns.lineplot(data=pas, x='Year', y='Trade Value (US$)', hue='Partner')
sns.set(rc={'figure.figsize':(11.7,8.27)})

plt.yscale('linear')
plt.xlabel('Tahun')
plt.ylabel('Nilai Ekspor')
plt.title('Ekspor  Indonesia 2001-2020')
plt.show()
```
