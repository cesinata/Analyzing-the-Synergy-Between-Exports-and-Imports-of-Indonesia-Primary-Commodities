```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_export_aug_21 = pd.DataFrame({
    'Persentase': [76.42, 17.01, 4.98, 1.59]},
    index = ['Industri', 'Tambang', 'Migas', 'Pertanian'])

df_export_aug_21
```

```python
df_export_aug_21.plot.pie(y = 'Persentase')

centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('Export Agustus 2021')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1))

# sebagai vector ke SVG
plt.savefig('export_aug_21.svg', dpi=300)
  
# Displaing Chart
plt.show()
```
