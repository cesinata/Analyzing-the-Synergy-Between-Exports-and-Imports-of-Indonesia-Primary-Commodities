###Mengimpor CSV file Comtrade untuk Data Impor Mesin
###tahun 2001-2020 ke bentuk Dataframe pandas

```python
df_1 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Mesin/comtrade.csv')
df_2 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Mesin/comtrade (1).csv')
df_3 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Mesin/comtrade (2).csv')
df_4 = pd.read_csv('/content/drive/MyDrive/CesAlFem SIC/Mesin/comtrade (3).csv')
```

###Menggabungkan, mengurutkan, dan menampilkan Dataframe dari 
###Data Impor tahun 2001-2020

```python
df_all = df_1.append([df_2, df_3, df_4], ignore_index=True).sort_values(by='Year',
                                                     ascending=True,
                                                     kind='quicksort')

df_all[['Year', 'Trade Flow', 'Reporter', 'Partner', 'Trade Value (US$)']]
```

###Menampilkan visualisasi dari Dataframe menggunakan Lineplot

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

df_all['Year'] = df_all['Year'].astype('string')

sns.lineplot(data=df_all, x='Year', y='Trade Value (US$)', lw=4, label='Mesin', color='#4848FF',ax=ax)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),fancybox=False, shadow=True, ncol=5,fontsize=12)
plt.grid(False)
plt.savefig('Impor Mesin Indonesia 2001-2020', transparent=True)

ax = plt.gca()

plt.yscale('linear')

plt.show()
```
