```python
df_agg = df_coffee.groupby('Partner ISO').agg({'Trade Value (US$)': 'sum'}).reset_index().rename(columns={'Trade Value (US$)':'Trade Value 2019'})
```

```python
ax = sns.barplot(y='Partner ISO', x='Trade Value 2019', data=df_agg_sorted, palette='pastel')
plt.show()
```

```python
years = range(2001, 2022)
```

```python
list(years)
```

```python
for y in years:
    df_coffee = df[ (df['Year'] == y) & (df['Trade Flow'] == 'M')  & (df['Commodity'].str.contains('coffee'))]
    if len(df_coffee) > 0:
        df_agg = df_coffee.groupby('Partner ISO').agg({'Trade Value (US$)': 'sum'}).reset_index().rename(columns={'Trade Value (US$)':'Trade Value '+str(y)})
        df_agg_sorted = df_agg.sort_values(by=['Trade Value '+str(y)], ascending=False)
        ax = sns.barplot(y='Partner ISO', x='Trade Value '+str(y), data=df_agg_sorted, palette='ocean')
        plt.show()
```
